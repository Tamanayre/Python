#Importando base de datos



































Creamoa un DataFrame(data) leyendo el archivo externo credit_card.csv

```python
import pandas as pd
data=pd.read_csv('credit_card.csv')
data
```

Creamos un diccionario con los nombre de las columnas de nuestro df como claves, y como valor los nombres que deseamos colocarles


```python
dict_col= {
    'LIMIT_BAL': 'Limite',
    'CHECKING_ACCOUNT': 'Ct_Corriente',
    'EDUCATION': 'Escolaridad',
    'MARRIAGE': 'Edo_Civil',
    'AGE': 'Edad',
    'BILL_AMT': 'Valor_Facturra',
    'PAY_AMT' : 'Valor_Pago',
    'DEFAULT': 'Morosos'
}
```

Realizamos el cambio de los nombres de las columnas

```python
tarjetas= data.rename(columns= dict_col)
tarjetas
```

Solicitamos los valores unicos para la clumna Ct_Coriente

```python
tarjetas.Ct_Corriente.unique()
```

REalizamos el cambio de los valores de la columna Ct_Corriente

```python
tarjetas.Ct_Corriente.map({'Yes':'SI','No':'NO'})
```

```python
tarjetas.Ct_Corriente= tarjetas.Ct_Corriente.map({'Yes':'SI','No':'NO'})
tarjetas
```

Verificamos los valores que puede contener la columna Escolaridad

```python
tarjetas.Escolaridad.unique()
```

Renombramos los valores que contiene la columna Escolaridad

```python
tarjetas.Escolaridad = tarjetas.Escolaridad.map({'2.University':'2.Universidad', '3.Graduate School':'3.Postgrado', '1.High School':'1.Colegio'})
tarjetas
```

Verificamos los valores de la columna Edo_Civil

```python
tarjetas.Edo_Civil.unique()
```

Cambiamos los nombres de los valores de la columna Edo_civil

```python
tarjetas.Edo_Civil= tarjetas.Edo_Civil.map({'Married':'Casado', 'Single':'Soltero'})
tarjetas
```
